## Step 2 - Fusing the data

Thorughout the expriment we measured different metrics per participant. Those are stored in seperate CSV files. Some of them were auto-generated by additional scripts, some were created by hand.  
In a first step we are going to fuse all individual CSV files, using the *pariticpant codename* as key. The codename consists of a colour and an animal.

 > We used a total of 4 control groups, *Red*, *Green*, *Blue*, *Yellow*, each starring 7 participants. This setup was chosen to reduce learning effects.
 
 The outcome of below cell is a new file: ```restify.csv```.  
 This file lists all data we ever process in the following, indexed by *participant codename*.

````python
import pandas as pd

print('Merging the individual CSV files...')

# Actually merge the CSV files
statscsv=pd.read_csv("stats.csv")
taskcsv=pd.read_csv("tasks.csv")
skillcsv=pd.read_csv("partitionskills.csv")
intermediate1=statscsv.merge(taskcsv,on="codename")
result=intermediate1.merge(skillcsv,on="codename")
result.to_csv("restify.csv", index=None)

# Print the result
print('Done!\n')
restifycsv = open("restify.csv", "r").read()
print(restifycsv)
````
